```python
from sklearn.ensemble import IsolationForest
import pandas as pd

mouse = pd.read_csv("mouse.txt", header = None, sep = " ", names = ["X1", "X2"])
mouse.head()
mouse.tail()
mouse.isnull().any() # pas de donn√©es manquantes
```

```python
ilf = IsolationForest(n_estimators=100, n_jobs=-1, verbose=2)
ilf.fit(mouse)
pred = ilf.predict(mouse)
```

```python
pred

# print(mouse["X1"].min(), mouse["X1"].max(), mouse["X2"].min(), mouse["X2"].max())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

xx, yy = np.meshgrid(np.linspace(0, 1, 50), np.linspace(0, 1, 50))
Z = ilf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.title("IsolationForest Micky Mouse")
plt.contourf(xx, yy, Z, cmap=plt.cm.Blues_r)

b1 = plt.scatter(mouse[:-10].loc[pred[:-10] == -1]["X1"] , mouse[0:-10].loc[pred[0:-10] == -1]["X2"], c='red',
                 s=20, edgecolor='k')
b2 = plt.scatter(mouse[:-10].loc[pred[:-10] == 1]["X1"], mouse[0:-10].loc[pred[0:-10] == 1]["X2"], c='green',
                 s=20, edgecolor='k')
c = plt.scatter(mouse.loc[490:, "X1"], mouse.loc[490:, "X2"], c='white',
                s=20, edgecolor='k')
plt.axis('tight')
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.legend([b1, b2, c],
           ["Type I error",
            "Regular point", "True positive"],
           loc="lower left")
plt.show()
```

```python
train = pd.read_csv("SMSSpamCollection.data", header=None, delimiter="\t", quoting=3, names = ["class", "text"])
print(train.shape)

print(train["text"][0])
train.head()
```

```python
import re
from nltk.corpus import stopwords

def sms_to_words(raw):
    letters_only = re.sub("[^a-zA-Z]", " ", raw)
    words = letters_only.lower().split()
    # In Python, searching a set is much faster than searching a list, so convert the stop words to a set.
    stops = set(stopwords.words("english"))
    # Enveler les "mots d'arr√™t"
    meaningful_words = [w for w in words if not w in stops]
    # Join the words back into one string separated by spaces and return it.
    return( " ".join( meaningful_words))
    # return(" ".join(words))
    
train["text"] = train["text"].map(sms_to_words)
```

```python
train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.decomposition import TruncatedSVD

vectorizer = CountVectorizer(analyzer = "word", tokenizer = None, preprocessor = None, stop_words = "english", max_features = 5000)
train_data_frequencies = vectorizer.fit_transform(train["text"]).toarray()
train_data_frequencies.shape

transformer = TfidfTransformer()
train_data_tfidf = transformer.fit_transform(train_data_frequencies)
train_data_tfidf[:5, :5].todense()

svd = TruncatedSVD(16)
train_data_truncated = svd.fit_transform(train_data_tfidf)
```

```python
(train["class"] == "spam").sum()/train.shape[0]
```

```python
ilf_sms = IsolationForest(n_estimators=100, n_jobs=-1, verbose=2)
ilf_sms.fit(train_data_tfidf)
pred_sms = ilf_sms.predict(train_data_tfidf)
pred_sms
```

```python
real_sms = train["class"].str.get_dummies()["spam"]

pred_sms[pred_sms == 1] = 0 # run only once !
pred_sms[pred_sms == -1] = 1 #run only once !
pred_sms
confusion_matrix(real_sms, pred_sms)
```

```python
print(real_sms[:15])
print(pred_sms[:15])
```

```python
# Cette √©tape prend 5 min.
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
    
pca = PCA(n_components=2, random_state=0)
tsne = TSNE(n_components=2, random_state=0, perplexity=50) # Ici 30 semble mieux que 50.

train_pca_2d = pca.fit_transform(train_data_truncated)
train_tsne_2d = tsne.fit_transform(train_data_truncated)

```

```python
print(train_pca_2d[:, 0].min(), train_pca_2d[:, 0].max(), train_tsne_2d[:, 0].min(), train_tsne_2d[:, 0].max())
len_x_pca = train_pca_2d[:, 0].max() - train_pca_2d[:, 0].min()
len_x_tsne = train_tsne_2d[:, 0].max() - train_tsne_2d[:, 0].min()

len_y_pca = train_pca_2d[:, 1].max() - train_pca_2d[:, 1].min()
len_y_tsne = train_tsne_2d[:, 1].max() - train_tsne_2d[:, 1].min()
```

```python
def drawConfusionMatrix(X_scores, y_real, y_pred, titre="IsolationForest", legend_loc = "upper right"):
    x_max = X_scores[:, 0].max()
    x_min = X_scores[:, 0].min()
    y_max = X_scores[:, 1].max()
    y_min = X_scores[:, 1].min()
    
    len_x = x_max - x_min
    len_y = y_max - y_min

    xx, yy = np.meshgrid(np.linspace(x_min - len_x / 10, x_max + len_x / 10, 50), 
                     np.linspace(y_min - len_y, y_max + len_y, 50))
    Z = ilf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.title(titre)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Blues_r)

    TN = plt.scatter(X_scores[:, 0][np.logical_and(y_pred == 0, y_real == 0)], X_scores[:, 1][np.logical_and(y_pred == 0, y_real == 0)], c='green',
                 s=20, edgecolor='k')
    TP = plt.scatter(X_scores[:, 0][np.logical_and(y_pred == 1, y_real == 1)], X_scores[:, 1][np.logical_and(y_pred == 1, y_real == 1)], c='white',
                 s=20, edgecolor='k')
    FN = plt.scatter(X_scores[:, 0][np.logical_and(y_pred == 0, y_real == 1)], X_scores[:, 1][np.logical_and(y_pred == 0, y_real == 1)], c='red',
                 s=20, edgecolor='k')
    FP = plt.scatter(X_scores[:, 0][np.logical_and(y_pred == 1, y_real == 0)], X_scores[:, 1][np.logical_and(y_pred == 1, y_real == 0)], c='orange',
                 s=20, edgecolor='k')
    plt.axis('tight')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.legend([TN, TP, FN, FP], ["True negative", "True positive", "False negative", "False positive"], loc=legend_loc)
    plt.show()
```

```python
drawConfusionMatrix(train_tsne_2d, real_sms, pred_sms)
```
